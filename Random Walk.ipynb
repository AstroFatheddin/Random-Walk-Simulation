 **1D Random Walk**

```python
from random import choice
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
N=100  #Number of Steps
M=5000 #Number of People
x1=0 #Start Point
a=1.0 
x=[x1]
x_sum =np.zeros(N)
x2_sum=np.zeros(N)
t = range(N)
for j in range (M):
    x1=0
    x=[x1]
    for i in range (N):
        s=choice([a,-a])
        x1=x1+s
        x.append(x1)
    for i in range (N):
        x_sum[i]=x_sum[i]+x[i]
        x2_sum[i]=x2_sum[i]+x[i]**2
x_avg=[]
x2_avg=[]
for i in range (N):
    x_avg.append(x_sum[i]/M)
    x2_avg.append(x2_sum[i]/M)
plt.xlabel('N')
plt.ylabel('avg')
plt.plot(t,x_avg , 'b-',label='x_avg')
plt.legend()
```

 **2D Random Walk**

```python
import numpy as np 
import pylab 
import random 
n = 1000 
x = np.zeros(n) 
y = np.zeros(n) 
direction=["NORTH","SOUTH","EAST","WEST"]
for i in range(1, n): 
    step = random.choice(direction)  
    if step == "EAST": 
        x[i] = x[i - 1] + 1
        y[i] = y[i - 1] 
    elif step == "WEST": 
        x[i] = x[i - 1] - 1
        y[i] = y[i - 1] 
    elif step == "NORTH": 
        x[i] = x[i - 1] 
        y[i] = y[i - 1] + 1
    else: 
        x[i] = x[i - 1] 
        y[i] = y[i - 1] - 1
pylab.title("Random Walk 2-D")
pylab.plot(x, y) 
pylab.show()
```

 **2D Self Avoiding Random Walk**

```python
import random
import matplotlib.pyplot as plt
%matplotlib inline 
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

n_walks=2
n_steps=35
walks=0
while walks< n_walks:
    sites=[]
    x=0 ; y=0
    sites.append([x,y])
    walk_succeeded=True
    
    for step in range(n_steps):
        direction=random.randrange(4)
        if direction==0:
            x += 1
        elif direction==1:
            y += 1
        elif direction==3:
            x -= 1
        else:
            y -= 1
        if sites.count([x,y])==1:
            visited=True
        else:
            visited=False
            
        if visited:
            walk_succeeded=False
            break
        else:
            sites.append([x,y])
    if walk_succeeded:
        walks += 1
        
plt.plot(sites)            
```

 **3D Random Walk**

```python
N =1500
R = (np.random.rand(N)*6).astype("int") 
x = np.zeros(N) 
y = np.zeros(N)
z = np.zeros(N)
x[ R==0 ] = -1; x[ R==1 ] = 1 
y[ R==2 ] = -1; y[ R==3 ] = 1
z[ R==4 ] = -1; z[ R==5 ] = 1
x = np.cumsum(x) 
y = np.cumsum(y)
z = np.cumsum(z)
plt.figure()
ax = plt.subplot(1,1,1, projection='3d')
ax.plot(x, y, z,alpha=0.9) 
ax.scatter(x[-1],y[-1],z[-1])
plt.show()
```
